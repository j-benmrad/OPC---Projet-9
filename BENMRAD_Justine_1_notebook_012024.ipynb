<div style="border: 1px solid #33A5B6; border-radius: 8px; padding: 20px; background-color: #F0F8FF;">
<h3 style="margin: auto; color: #33A5B6;">7.2 - COURBE DE LORENZ</h3>
</div>

```python
# Identification du CA généré par client
ca_client = df_BtoC.groupby(['client_id'])['price'].sum().reset_index()
ca_client.sort_values(by='price', inplace=True)
display(ca_client.head())
```

```python
# Calcul de la courbe de Lorenz

## Selection de la colonne prix
dep = ca_client["price"].values

## Calcul de Lorenz
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0], lorenz)  # Ajoute 0 en premier élément

## Définition des axes du graphique
plt.figure(figsize=(12, 8))
xaxis = np.linspace(0 - 1/n, 1 + 1/n, n + 1)

## Définition du traçage de la courbe
plt.plot(xaxis, lorenz, drawstyle='steps-post')
plt.plot([0, 1], [0, 1], color='lightcoral', linestyle='solid', label="Courbe d'égalité parfaite")

## Affichage du graphique
plt.show()
```

```python
# Calcul de l'indice de Gini

## Calcul de l'air sous la courbe (AUC)
n = len(lorenz)  # Nombre de clients (taille du tableau lorenz)
AUC = (lorenz.sum()-lorenz[-1])/n  # Calcul de l'aire sous la courbe

## Calcul de la surface (S)
S = 0.5 - AUC

## Calcul de l'indice de Gini
gini = (2*S).round(3)
print(f"L'indice de Gini est égal à {gini}")
```

```python
## CONCLUSION

# Courbe de Lorenz : outil graphique utilisé pour représenter la répartition d'une variable quantitative (ici, le prix) au sein d'une population (ici, les clients).
# Axe des abscisses : les clients classés par ordre croissant en partant de celui qui dépense le moins
# Axe des ordonnées : part cumulée des dépenses détenues par un % de clients
# Droite diagonale : droite d'égalité parfaite
# Courbe : inégalités 
# >>> Présence d'inégalités dans les dépenses


# Indice de Gini: mesure quantitative de l'inégalité, entre 0 (égalité parfaite) et 1 (inégalité maximale)
# >>> L’indice de Gini de 0.398 indique une inégalité modérée dans les dépenses
# >>> Une petite partie des clients représentent une part importante des dépenses
```

<h3 style="margin: auto; color: #33A5B6;">CORRELATION 1 : GENRE DU CLIENT vs CATEGORIES DE LIVRES ACHETE</h3>
</div>

```python
# Affichage du prix par genre et par categorie
sex_categ_count = df_BtoC.groupby(['sex', 'categ'])['id_prod'].count().unstack().fillna(0)
display(sex_categ_count.head())
```

```python
# Création d'un graphique en heat map
import matplotlib.pyplot as plt
import seaborn as sns

data = sex_categ_count.fillna(0)
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt="d", cmap="Greens", vmin=0)
plt.xlabel("Catégorie de livre", fontsize=16)
plt.ylabel("Genre", fontsize=16)
plt.title("Nombre de livres vendus, par genre du client, et par catégorie de livre", fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
## INTERPRETATION VISUELLES DU GRAPHIQUE
# Les F achètent plus que les H, peu importe la catégorie de livres
# La heatmap ne montre pas de corrélation particulière


## CHOIX DANS LA DEMONSTRATION DE LA CORRELATION : TEST DU CHI2 + p_value + V de Cramer
# Test du Chi2 : association entre 2 variables qualitatives / teste d'indépendance entre ces 2 variables
# P_value : probabilité / plausibilité de H0 au vu des données. Seuil de 5%
# V de Cramer : évalue l'intensité de l'association entre 2 variables quali ; variation entre 0 et 1


## CALCUL
from scipy.stats import chi2_contingency
# Création d'une table de contingence
contingency_table = pd.crosstab(df_BtoC['sex'], df_BtoC['categ'])
# Utilisation du test Chi2
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table.values)
# Calcul V de Cramer
n_sc = len(df_BtoC)
minDim_sc = min(sex_categ_count.shape)-1
V_sc = np.sqrt((chi2_stat/n_sc) / minDim_sc)
# Affichages
print("Résultat du test statistique Chi2 :", chi2_stat)
print("Résultat du test p_value :", p_value)
print("Le V de Cramer est de:", V_sc)
# Interprétation du résultat
if p_value < 0.05:
    print("Il y a une corrélation significative entre le genre du client et la catégorie de livre qu'il achète (p-value < 0.05).")
else:
    print("Il n y a pas de corrélation significative entre le genre du client et la catégorie de livre qu'il achète (p-value >= 0.05).")
    
    
    
## INTERPRETATION
# Test du Chi2 >>> corrélation OUI
# p_value >>> confirme la corrélation
# V de Cramer >>> corrélation très faible


## CONCLUSION
# Corrélation faible, peut êre due à la grande taille de l'échantillon   
```

<div style="border: 1px solid #33A5B6; border-radius: 8px; padding: 20px; background-color: #F0F8FF;">
<h3 style="margin: auto; color: #33A5B6;">CORRELATION 2 : AGE DU CLIENT vs MONTANT TOTAL DES ACHATS</h3>
</div>

```python
# Création d'une colonne pour avoir l'âge du client en 2024
df_BtoC.loc[:, 'age'] = 2024 - df_BtoC['birth']
display(df_BtoC.head())
```

```python
# Affichage du montant dépensé selon l'âge
age_price = df_BtoC.groupby('age')['price'].sum().reset_index()
display(age_price.head())
```

```python
import matplotlib.pyplot as plt

# Création d'un graphique en nuage de points
plt.scatter(age_price['age'], age_price['price'], color='olive')
plt.xlabel('Âge', fontsize=16)
plt.ylabel('Montant total des achats', fontsize=16)
plt.title('Montant total des achats en fonction de l\'âge', fontsize=20)
plt.show()
```

```python
## INTERPRETATION VISUELLE DU NUAGE DE POINTS
# 2 phases : asc jusque 45 ans puis desc
# Relation non linéaire
# Corrélation négative entre âge et montant total des achats ?


## CHOIX DANS LA DEMONSTRATION DE LA CORRELATION : COEFFICIENT DE SPEARMAN
# Relation monotone entre 2 variables
# Relation non linéaire
# -1 corrélation négative forte ; 0 pas de corrélation ; 1 corrélation positive forte


## CALCUL DU COEFFICIENT DE SPEARMAN ET DE SA P_VALUE
# Calcul
import scipy.stats as stats
spearman_coeff, p_value = stats.spearmanr(age_price['age'], age_price['price'])
# Affichage des résultats
print("Coefficient de corrélation de Spearman:", spearman_coeff)
print("p-value:", p_value)

## INTERPRETATION
# Le coefficient de Spearman est de -0.874 : un coefficient proche de -1 indique une corrélation négative forte !
# Au fur et à mesure que l'âge augmente, le montant total des achats diminue de manière significative.
# la p_valeur est de 5.956e-25 : la p-valeur est extrêmement faible (quasi nulle), ce qui signifie que nous pouvons rejeter l'hypothèse nulle selon laquelle il n'y a pas de corrélation entre les deux variables.
# II existe une corrélation statistiquement significative entre ces deux variables.

## CONCLUSION
# Les résultats de l'analyse indiquent qu'il existe une corrélation négative très forte et statistiquement significative entre l'âge et le montant total des achats.
# C'est à dire que, plus une personne est jeune, plus elle dépense, et que plus une personne est âgée, moins elle dépense.
```

```python
# Affichage du montant dépensé, par mois, selon l'âge
age_frequence = df_BtoC.groupby(['age', 'year_month'])['id_prod'].count().reset_index()
# Renommage de la colonne
age_frequence = age_frequence.rename(columns={'id_prod': 'nbr_products'})
display(age_frequence.head())
```

```python
# Création d'une fonction créer des tranches d'âge
def tranche_age(age): 
    if 20 <= age < 35:
        return "20-34 ans"
    elif 35 <= age < 50:
        return "35-49 ans"
    elif 50 <= age < 65:
        return "50-64 ans"
    elif 65 <= age < 80:
        return "65-79 ans"
    else:
        return "80 ans et plus"

# Application de la fonction à la colonne 'age' et création d'une nouvelle colonne 'tranche_age'
age_frequence['tranche_age'] = age_frequence['age'].apply(tranche_age)
display(age_frequence.head())
```

```python
# Regroupement pour affichage du montant dépensé, par mois, selon la tranche d'âge
age_frequence = age_frequence.groupby(['tranche_age', 'year_month'])['nbr_products'].sum().reset_index()
display(age_frequence.head())
```

```python
# Création d'un graphique en ligne avec plusieurs lignes (une par tranche d'âge)

# Création des couleurs
colors = ['greenyellow', 'gold', 'coral', 'mediumpurple', 'skyblue']

# Create a line chart with different colors for each tranche d'age
plt.figure(figsize=(22, 6))
for tranche, color in zip(age_frequence['tranche_age'].unique(), colors):
    data = age_frequence[age_frequence['tranche_age'] == tranche]
    plt.plot(data['year_month'], data['nbr_products'], label=tranche, color=color)

plt.xlabel('Mois', fontsize=16)
plt.ylabel('Nombre de livres achetés', fontsize=16)
plt.title('Évolution du nombre de livres achetés par tranche d\'âge', fontsize=20)
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Affichage de la taille moyen du panier en fonction de l'âge du client
age_panier = df_BtoC.groupby('age')['price'].mean().reset_index()
display(age_panier.head())
```

```python
# Création d'un graphique en line charts
plt.figure(figsize=(8, 5))
plt.plot(age_panier['age'], age_panier['price'], marker='o', linestyle='-', color='olive')
plt.xlabel('Age', fontsize=16)
plt.xlim(left=0)
plt.ylabel('Taille du panier moyen', fontsize=16)
plt.ylim(bottom=0)
plt.title('Taille du panier moyen en fonction de l\'âge du client', fontsize=20)
plt.grid(True)
plt.show()
```

```python
## INTERPRETATION VISUELLE DU GRAPHIQUE
# 3 plateaux
# 20/30 ans = grande variabilité dans la taille du panier moyen
# A partir de 30 ans la taille du panier moyen s'éffondre
# A partir de 50 ans, la taille du panier moyen se stabilise

## CHOIX DANS LA DEMONSTRATION DE LA CORRELATION : LA REGRESSION PAR SPLINES
# Relation non-linéaire complexe
# Relation avec des changements brusques

## CALCUL
import statsmodels.api as sm
from statsmodels.gam.api import GLMGam, BSplines
# Création de la matrice de splines
bs = BSplines(age_panier['age'], df=7, degree=3)
# Ajustement du modèle GAM
gam_model = GLMGam.from_formula('price ~ age', data=age_panier, smoother=bs).fit()
# Affichage du résumé du modèle
print(gam_model.summary())

## INTERPRETATION
# Présence de splines ==> relation non linéaire entre age et taille du panier moyen
# Pas de constance
# si splines négatif alors diminution de la taille du panier moyen autour de l'âge correspondant à ce splines

## CONCLUSION
# Relation non linéaire
# Variations importantes en fonctions de l'âge
```

```python
# Visualisation des résultats
import matplotlib.pyplot as plt

# Prédication des valeurs pour tous les âges
age_panier['predicted'] = gam_model.predict()

# Affichage du graphique
plt.scatter(age_panier['age'], age_panier['price'], label='Données', color='olive')
plt.plot(age_panier['age'], age_panier['predicted'], color='red', label='Régression par splines')
plt.title('Visualisation de la régression par splines', fontsize=16)
plt.xlabel('Âge', fontsize=16)
plt.xlim(left=0)
plt.ylabel('Panier moyen', fontsize=20)
plt.ylim(bottom=0)
plt.legend()
plt.show()
```

```python
## INTERPRETATION VISUELLE DU GRAPHIQUE
# variations dans le nombre de livres achetés selon les âge et les catégories de livre
 
    
## CHOIX DANS LA DEMONSTRATION DE LA CORRELATION : TEST ANOVA
# Moyenne du nombre de livres achetés, différente selon âge et catégorie ?
# Interaction entre les 2 variables ?


## Calcul
# importation des librairies
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
# Création d'un dataframe 'long'
df_anova = df_BtoC.groupby(['client_id', 'categ', 'age'])['id_prod'].count().reset_index()
df_anova = df_anova.rename(columns={'id_prod': 'nbr_products'})
display(df_anova.head())
# Calcul du modèle ANOVA avec interaction entre l'âge et la catégorie
model = ols('nbr_products ~ C(age) * C(categ)', data=df_anova).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)


## INTERPRETATION
# effet significatif de l'âge sur la catégorie de livre achetés
# effet significatif de la catégorie de livre sur l'âge
# effet significatif de l'interaction entre l'âge et la catégorie


## CONCLUSION
# L'effet de la catégorie sur le nombre de livres achetés dépend de l'âge, et vice versa.
```
